```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the results
stationary = pd.read_csv('submissions_round1.csv')
non_stationary = pd.read_csv('submissions_round2.csv')

# Combine them for comparison
stationary['environment'] = 'Stationary Casino'
non_stationary['environment'] = 'Non-Stationary Casino'
combined = pd.concat([stationary, non_stationary])

# Bar chart comparison
plt.figure(figsize=(6,4))
plt.bar(combined['environment'], combined['total_reward'], color=['skyblue', 'lightcoral'])
plt.title('Exploration–Exploitation Trade-off\n(Adhitya Kondeti – 8997046)')
plt.ylabel('Total Reward Earned')
plt.xlabel('Environment Type')
plt.grid(axis='y')
plt.show()

# Print numerical comparison
print(combined[['environment', 'strategy', 'details', 'total_reward']])
```

## Reflection and Analysis

**1. Observation:**  
The ε-greedy strategy performed better in the **Non-Stationary Casino (≈ 2557 rewards)** compared with the **Stationary Casino (≈ 1862)**.  
This indicates that a constant learning rate (α = 0.1) helped the agent adapt to changing reward probabilities.

**2. Why ε Matters:**  
ε controls how often the agent explores new actions versus exploiting the best-known one.  
A moderate value like 0.1 allows enough exploration to avoid being stuck on sub-optimal arms while still exploiting good ones.

**3. Why Constant α Helps in Non-Stationary Settings:**  
In environments where reward distributions shift over time, a constant α ensures that recent experiences have stronger influence, helping the model quickly track new reward trends.

**4. Takeaway:**  
Exploration–exploitation balance is environment-dependent.  
Non-stationary settings benefit from adaptive or constant learning rates, while stationary cases reward stable convergence.
